```python
import pandas as pd
```

```python
countries_to_filter = ["Iran", "Vietnam", "Egypt", "Namibia",
                      "Ireland", "Hungary", "Canada", "Argentina"]
```

```python
life_exp_df = pd.read_csv("data/gapminder.csv")

life_exp_df.head(3)
```

```python
continent_df = life_exp_df[life_exp_df.country.isin(countries_to_filter)].iloc[:, [0, 1, 2, 3, 4]]
continent_df.head(3)
```

```python
life_exp_proj = pd.read_csv("data/life_expectancy_at_birth_with_projections.csv")
life_exp_proj = life_exp_proj[life_exp_proj.country.isin(countries_to_filter)]
life_exp_proj.head(3)
```

```python
life_exp_proj_new = life_exp_proj.merge(continent_df)

life_exp_proj_new.head(3)
```

```python
popn_proj = pd.read_csv("data/total_population_with_projections.csv")

popn_proj.head(3)
```

```python
extra_years = range(2008, 2101)

life_exp_df = continent_df.copy()
```

```python
for idx, row in life_exp_proj_new.iterrows():
    for year in extra_years:
        life_exp_df = life_exp_df.append(pd.DataFrame({"country": row.country, "continent": row.continent, 
                            "year": year, "lifeExp": row[str(year)], 
                            "pop": popn_proj[popn_proj.country==row.country][str(year)]}), 
                           ignore_index=True)
```

```python
life_exp_df.to_csv("data/exp_popn_proj.csv")
```
