```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('seaborn')
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  columns=list('ABCD'))

df = df.cumsum()

plt.figure()

df.plot()
plt.show()
```
