## Exploring a Gapminder dataset

## 1) Read the CSV data

```python
import pandas as pd
```

```python
gapminder_df = pd.read_csv('data/gapminder.csv')

countries_to_filter = ["Iran", "Vietnam", "Egypt", "Namibia",
                      "Ireland", "Hungary", "Canada", "Argentina"]
```

### 2) Plot

```python
data = gapminder_df[gapminder_df.country.isin(countries_to_filter)]

alt.Chart(data.reset_index(), width=800, height=400).mark_line().encode(
    x='year:N',
    y='lifeExp',
    color=alt.Color('country', legend=alt.Legend(orient="bottom")),
    tooltip=['year', 'country', 'continent', 'lifeExp']
).properties(
    title='Evolution of life expectancy in 8 countries'
).configure_axis(
    grid=False
)
```
